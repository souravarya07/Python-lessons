```python
#Style and Labelling
x = list(range(1,100,3))
y = [t**2 for t in x]
plt.plot(x,y, 'r-o', markersize=8)
# plt.plot(y, 'g-o',markersize=8)
```

```python
import numpy as np
x = np.arange(1,102,5)
y = x**2 #Numpy is a vectorized library

plt.title('Square')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.xlim(1,101)
plt.ylim(1,10000)

plt.plot(x,y)
plt.show()
```

```python
#Plotting multiple graphs
x = np.arange(1,10)
y = x**2
y1 = x**3

plt.title('Square Vs. Cube')
plt.plot(x,y,label='Square')
plt.plot(x,y1,label='Cube')
plt.legend()
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.show()

```

```python
x = np.arange(1,10)
y = x**2
y1 = x + 50

plt.title('Square Vs. Sum')
plt.plot(x,y,label='Square')
plt.plot(x,y1,label='Sum')
plt.legend()
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.text(1.8,8,'(2,4)')
plt.show()
```

```python
import math
x = np.arange(1,101,5)

def custom_func(x):
    return (x+50)**2/2

vectorized_log = np.vectorize(custom_func)
y = vectorized_log(x)
plt.plot(x,y, 'g-*')
plt.show()
```

```python
#Bar Graphs
plt.bar([1,2,3],[10,20,8])
plt.show()
```

```python
width = 0.2
x = np.array([1,2,3])
y_1 = np.array([10,20,3])
y_2 = np.array([4,12,9])
plt.bar(x-(width/2), y_1, width = width,color='r', label='crime-rate')
plt.bar(x+(width/2),y_2, width = width,color = 'g' ,label='murder-rate')
plt.legend()
plt.show()
```

```python
#Histogram
vals = np.random.rand(100)
plt.hist(vals,20)
plt.show()
```

```python
mu = 1000
sigma = 50
vals = np.random.normal(mu, sigma, 100)
plt.plot(vals,'g-o')
```

```python
#Normal distribution
mu = 1000
sigma = 50
vals = np.random.normal(mu, sigma, 10000)
counts, bins,_ = plt.hist(vals, bins = 100)
plt.show()
```

```python
#Scatter Plot
x = np.random.rand(100)
y = np.random.rand(100)

x1 = np.random.rand(100)+0.2 
y1 = np.random.rand(100)

plt.scatter(x, y)
plt.scatter(x1, y1)
plt.show()
```

```python
def sigmoid(x):
    return math.tanh(x)

vectorized_sigmoid = np.vectorize(sigmoid)

x = np.arange(-10, 10, 0.1)
y = vectorized_sigmoid(x)

plt.plot(x, y)
plt.show()
```
