```python
#Missing values
disease = pd.read_csv("disease.csv")
disease.head()
```

```python
from numpy import NaN, NAN, nan
pd.isnull(nan)
```

```python
#If you have just one or very few rows with missing data, you can simply remove those rows.
disease['Cases_Guinea'].value_counts(dropna=False).head()
```

```python
disease['Cases_Guinea'].unique()
```

```python
disease['Cases_Guinea'].nunique(dropna=False)
```

```python
#Replace the value with mean
avg = disease['Cases_Guinea'].mean()
disease['Cases_Guinea'] = disease['Cases_Guinea'].replace(to_replace=NaN, value=avg)
disease.head()
```

```python
#Backward and Forward fill approach 
disease = disease.fillna(method="ffill")
disease = disease.fillna(method="bfill")
disease.head()
```

```python
#Restructuring the Data
weather = pd.read_csv("weather.csv")
weather.head()
```

```python
weather_melt = pd.melt(weather, id_vars=['year', 'month', 'element'], var_name="day", value_name="temp")
weather_melt.head()
```

```python
weather_tidy = weather_melt.pivot_table(index=['year', 'month', 'day'], columns= 'element', values='temp', dropna=False).reset_index()
weather_tidy.head()
```

```python
#Write CSV files 
weather.to_csv("weather_tidy.csv", sep=',')
```

```python
#Apply a function on multiple columns or rows
def mean_func(x):
    return sum(x)/len(x)

weather_temp = weather_tidy[['max', 'min']]
weather_temp.apply(mean_func, axis=0) #axis would be 0 because we want to collapse rows
```

```python
#Merging
users = pd.DataFrame({'id':[1,2,3], 'name':['A', 'B', 'C']})
msgs = pd.DataFrame({'user_id':[1,1,3], 'msg':['hello', 'bye', 'hi']})
users.merge(msgs, left_on='id', right_on='user_id') #This is known as Inner join
```
