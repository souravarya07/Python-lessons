```python
le = df['lifeExp'] #It works because life expectancy is a key
m = le.mean()
m 
```

```python
le > 50
```

```python
#Filtering Methods
df.loc[df['lifeExp']>80] #Returns rows with lifeExp greater than 80
```

```python
df.loc[df.lifeExp > m,['Country','lifeExp']] #returns rows with life expectancy greater than mean along with country name.
```

```python
#Multiple Conditions - Similar to SQL Queries
df.loc[(df.lifeExp < 60) & (df.lifeExp > 30)].head(10) 
```

```python
#Multiple conditions on Rows with selection of certain columns
df.loc[(df.lifeExp < 30) | (df.lifeExp < 62), ['year', 'Country', 'lifeExp']].head(10)
```

```python
#Filter based on Strings
df.loc[df.Country == "Zimbabwe"]
```

```python
df.loc[(df.Country=="Kenya")| (df.Country=="Egypt")]
```

```python
df.loc[df.Country > "Nigeria"].head(30)#Strings > Nigeria
```

```python
#Grouping
df.groupby('year')['lifeExp'].mean().reset_index() #Gives the mean for each group, can specify multiple columns here.
```

```python
df.groupby('year')['gdpPerCap'].mean().reset_index()
```

```python
#Sorting
df.sort_values(['year', 'lifeExp']).reset_index().head(10)
df.sort_values(['year', 'lifeExp'], ascending=False).head(10)
```
